```python
import numpy as np
import pandas as pd
```

```python
# load datasets
nodes = pd.read_csv('http://localhost:8888/tree/data/Nodes.csv')

# Node 1 data
# Latitudes
n1_lats = np.round([nodes['LAT1'][0], nodes['LAT1'][1], nodes['LAT1'][2]], 6)

# Longitudes
n1_lngs = np.round([nodes['LON1'][0], nodes['LON1'][1],nodes['LON1'][2]], 6)

# Node 2 data
# Latitudes
n2_lats = np.round([nodes['LAT2'][0], nodes['LAT2'][1], nodes['LAT2'][2]], 6)

# Longitudes
n2_lngs = np.round([nodes['LON2'][0], nodes['LON2'][1], nodes['LON2'][2]], 6)

# Node 3 data
# Latitudes
n3_lats = np.round([nodes['LAT3'][0], nodes['LAT3'][1], nodes['LAT3'][2]], 6)

# Longitudes
n3_lngs = np.round([nodes['LON3'][0], nodes['LON3'][1], nodes['LON3'][2]], 6)

# Node 4 data
# Latitudes
n4_lats = np.round([nodes['LAT4'][0], nodes['LAT4'][1], nodes['LAT4'][2]], 6)

# Longitudes
n4_lngs = np.round([nodes['LON4'][0], nodes['LON4'][1], nodes['LON4'][2]], 6)
```

```python
import gmplot

# Singapore lat/lon 1.290270, 103.851959
gmap = gmplot.GoogleMapPlotter(1.290270, 103.851959, 12)

# plot Node 1
gmap.plot(n1_lats, n1_lngs, 'cornflowerblue', edge_width=5) # lines
gmap.scatter(n1_lats, n1_lngs, 'k', size=100, marker=False) # points

# plot Node 2
gmap.plot(n2_lats, n2_lngs, 'cornflowerblue', edge_width=5) # lines
gmap.scatter(n2_lats, n2_lngs, 'k', size=100, marker=False) # points

# plot Node 3
gmap.plot(n3_lats, n3_lngs, 'cornflowerblue', edge_width=5) # lines
gmap.scatter(n3_lats, n3_lngs, 'k', size=100, marker=False) # points

# plot Node 4
gmap.plot(n4_lats, n4_lngs, 'cornflowerblue', edge_width=5) # lines
gmap.scatter(n4_lats, n4_lngs, 'k', size=100, marker=False) # points

# gmap.scatter(marker_lats, marker_lngs, 'k', marker=True)
# gmap.heatmap(heat_lats, heat_lngs)

gmap.draw("mymap.html")
```

```python
# load prediction 1
prediction = pd.read_csv('http://localhost:8888/tree/data/found.csv')
prediction
```

```python
# plot predicted Node 4 location
pred_n4_lats = [n4_lats[2], prediction['LAT4'][3]]
pred_n4_lngs = [n4_lngs[2], prediction['LON4'][3]]

gmap.plot(pred_n4_lats, pred_n4_lngs, 'green', edge_width=5) # lines
gmap.scatter(pred_n4_lats, pred_n4_lngs, 'c', size=25, marker=False) # points

gmap.draw("mymap.html")
```
